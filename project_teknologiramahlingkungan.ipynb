```python
# Menghitung jumlah transaksi per vendor
vendor_counts = df_clean['Nama Vendor'].value_counts()

# Mengambil 10 vendor teratas
top_vendors = vendor_counts.head(10)

plt.figure(figsize=(8, 6))
ax = sns.barplot(x=top_vendors.index, y=top_vendors.values, palette='flare')

# Menambahkan angka pada batang dengan format integer
for p in ax.patches:
    height = int(p.get_height())  # Mengubah tinggi batang menjadi integer
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.title('5 Vendor Teratas Berdasarkan Jumlah Transaksi')
plt.xlabel('Nama Vendor')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Memutar label x agar lebih mudah dibaca
plt.show()
```
